```python
from statistics import *
#Lista de calificaciones

grupoa = [70, 70, 65, 90, 100, 95, 45, 80, 73, 20, 100, 96]
print("Calificaciones")
print(grupoa)

type(grupoa)
#Calculo del promedio
promedio= sum(grupoa)/len(grupoa)
print("Promedio de calificaciones")
print(promedio)

promedio2= mean(grupoa)
print(promedio2)

mediana= median(grupoa)
print("Mediana de las calificaciones")
print(mediana)

moda= mode(grupoa)
print("Moda de las calificaciones")
print(moda)

moda2= multimode(grupoa)
print(moda2)

print("\t")
#Lista de tiempo
tiempo100mts = [15.10, 15.20,14.69,13.27,22.15,18.71,19.15,20.65,15.10,17.20]
#Impresion de los tiempos
print(tiempo100mts)

#Promedio
print("El tiempo promedio de los corredoreses=>", mean(tiempo100mts))

#Mediana
print("El tiempo promedio de los corredoreses=>", median(tiempo100mts))

#Moda
multimodas = multimode(tiempo100mts)
print("El tiempo promedio de los corredoreses=>", multimodas)
```

```python
import random
fabrica1 = []
for i in range (100):
  numero_aleatorio = round( random.uniform(50,100),2)
  fabrica1.append(numero_aleatorio)
print(fabrica1)
```

```python
import random
fabrica2 = []
for i in range (100):
  numero_aleatorio = round( random.uniform(50,100),2)
  fabrica2.append(numero_aleatorio)
print(fabrica2)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(fabrica1)
plt.title("Produccion diaria de la fabrica 1")
```

```python
plt.hist(fabrica2)
plt.title("Produccion diaria de la fabrica 2")
```

```python

media1 = mean(fabrica1)
mediana1 = median(fabrica1)
moda1 = mode(fabrica1)
print("Produccion diaria de la fabrica 1")
print("La media es = ",media1)
print("La mediana es =",mediana1)
print("La moda es = ",moda1)
```

```python
media2 = mean(fabrica2)
mediana2 = median(fabrica2)
moda2 = mode(fabrica2)
print("Produccion diaria de la fabrica 2")
print("La media es = ",media2)
print("La mediana es =",mediana2)
print("La moda es = ",moda2)
```

```python
flg, (ax1,ax2) = plt.subplots(nrows=1,ncols=2, figsize=(10,5))
ax1.hist(fabrica1)

ax2.hist(fabrica2)
plt.show()
```

```python
plt.hist(fabrica1, bins=10, alpha=0.2, color='b', edgecolor='black')
plt.axvline(media1, color='b', linestyle='dashed', linewidth=2, label='Media')
plt.axvline(mediana1, color='r', linestyle='dashed', linewidth=2, label='Mediana')
plt.axvline(moda1, color='y', linestyle='dashed', linewidth=2, label='Moda')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title("Produccion diaria de la fabrica 1")
plt.legend()
plt.show()
```
