```python
import pandas as pd
g1 = g1= pd.read_csv("grupo1.csv")
g1.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(6,4))
sns.histplot(g1["Calificacion"], bins=10, kde=True, color='red')
plt.title("Grupo 1 - Histograma de Calificaciones")
plt.xlabel("Calificacion")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from scipy.stats import pearsonr
from sklearn.metrics import r2_score

modelo1 = LinearRegression()
modelo1.fit(g1[['HorasEstudio']], g1['Calificacion'])

#Pendiente e intercepto
pendiente1 = modelo1.coef_[0]
intercepto1 = modelo1.intercept_

print(f"Ecuación de la recta: y = {pendiente1:.4f} x + {intercepto1:.4f}")

#coeficiente de correlación de Pearson
coeficienteDeCorrelacion, _ = pearsonr(g1['HorasEstudio'], g1['Calificacion'])
print(f"Coeficiente de correlación de Pearson: r = {coeficienteDeCorrelacion:.4f}")

#Determinacion

y_pred = modelo.predict(g1[['HorasEstudio']])
r2_1 = r2_score(g1['Calificacion'], y_pred)
print(f"Coeficiente de determinación (R²): {r2_1:.4f}")
```

```python
#Grafica scatter y linea
plt.figure(figsize=(8, 6))
X = g1[['HorasEstudio']]
y = g1['Calificacion']
plt.scatter(X, y, color='blue', label='Datos reales')
plt.plot(X, modelo1.predict(X), color='red', linewidth=2, label='Línea de regresión')
plt.title('Relación entre Horas de Estudio y Calificación')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificación')
plt.legend()
plt.grid(True)
plt.show()
```

**1.- ¿Explicar cuál es la interpretación del coeficiente de regresión en términos de la relación entre el tiempo de estudio y las calificaciones finales?**

En el grupo numero 1 la calificacion aumenta 2.47 por cada hora que un alumno estudia.

**¿Cómo se interpreta el coeficiente de correlación en términos de la fuerza y dirección de la relación entre las variables?**

Se interpreta como una correlacion positiva, ya que el valor de esta fue 0.95

**¿El modelo de regresión lineal es estadísticamente significativo?**

Si es significativo el modelo explica el 91% de las variaciones en las calificaciones.

**¿Qué tan confiable es el modelo de regresión para predecir las calificaciones finales de los estudiantes basado en el tiempo de estudio?**

 Por el R² tan alto y la correlación fuerte que existe, este modelo nos permite hacer buenas predicciones sobre la calificación que podría obtener un alumno en funcion de sus horas de estudio
