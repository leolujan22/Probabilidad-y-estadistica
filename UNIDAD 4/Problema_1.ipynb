```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
media = 70
desviacion = 4
totalEstudiantes = 500

min = 68
max = 72

x = np.linspace(media - 4 * desviacion, media + 4 * desviacion, 1000)
y = norm.pdf(x, media, desviacion)

entre = norm.cdf(max, media, desviacion) - norm.cdf(min, media, desviacion)
estudiantesEntre = entre * totalEstudiantes

plt.figure(figsize=(6, 4))
plt.plot(x, y, 'k', linewidth=2, label='Distribucin Normal')
plt.fill_between(x, y, where=(x >= min) & (x <= max), color='skyblue', alpha=0.5, label='Entre 68 y 72 kg')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de Probabilidad')
plt.title('Distribucin Normal: Estudiantes entre 68 y 72 kg')
plt.legend()
plt.grid(True)
plt.show()

print("Probabilidad entre 68 y 72 kg:", entre)
print("Nmero aproximado de estudiantes entre 68 y 72 kg:", estudiantesEntre)
```
